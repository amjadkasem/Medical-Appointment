```python
# Generate example data
np.random.seed(42)
data = pd.DataFrame({
    'scheduledday': np.random.choice(['2015'], 1000),
    'appointmentday': np.random.choice(['2016-05-16'], 1000),
    'hypertension': np.random.choice([0, 1], 1000, p=[0.7, 0.3]),
    'age': np.random.randint(0, 120, 1000),
    'diabetes': np.random.choice([0, 1], 1000, p=[0.8, 0.2]),
    'handicap': np.random.choice([0, 1], 1000, p=[0.9, 0.1]),
    'sms_received': np.random.choice([0, 1], 1000, p=[0.6, 0.4]),
})
# Set up subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Distribution of Variables in the Dataset', fontsize=16)
# Plot 1: Scheduled Appointments
sns.countplot(data=data, x='scheduledday', ax=axes[0, 0])
axes[0, 0].set_title('Scheduled Appointments (2015)')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Count')
# Plot 2: Actual Appointments
sns.countplot(data=data, x='appointmentday', ax=axes[0, 1])
axes[0, 1].set_title('Actual Appointments (2016-05-16)')
axes[0, 1].set_xlabel('Date')
axes[0, 1].set_ylabel('Count')
# Plot 3: Hypertension
sns.countplot(data=data, x='hypertension', ax=axes[0, 2])
axes[0, 2].set_title('Hypertension Distribution')
axes[0, 2].set_xlabel('Status (0: No, 1: Yes)')
axes[0, 2].set_ylabel('Count')
# Plot 4: Age
sns.histplot(data=data, x='age', bins=20, ax=axes[1, 0])
axes[1, 0].set_title('Age Distribution')
axes[1, 0].set_xlabel('Age (years)')
axes[1, 0].set_ylabel('Count')
# Plot 5: Diabetes
sns.countplot(data=data, x='diabetes', ax=axes[1, 1])
axes[1, 1].set_title('Diabetes Distribution')
axes[1, 1].set_xlabel('Status (0: No, 1: Yes)')
axes[1, 1].set_ylabel('Count')
# Plot 6: Handicap
sns.countplot(data=data, x='handicap', ax=axes[1, 2])
axes[1, 2].set_title('Handicap Distribution')
axes[1, 2].set_xlabel('Status (0: No, 1: Yes)')
axes[1, 2].set_ylabel('Count')
# Plot 7: SMS Received
sns.countplot(data=data, x='sms_received', ax=axes[2, 0])
axes[2, 0].set_title('SMS Receipt Status')
axes[2, 0].set_xlabel('Status (0: No, 1: Yes)')
axes[2, 0].set_ylabel('Count')
# Hide empty subplots
axes[2, 1].axis('off')
axes[2, 2].axis('off')
plt.tight_layout()
plt.show()  # <-- THIS LINE IS CRUCIAL TO DISPLAY PLOTS
```

```python
# Sample data (replace with your actual data)
gender = ['Female', 'Male']
show = [50000, 45000]       # Patients who showed up
no_show = [10000, 15000]    # Patients who didn't show
# Color palette
colors = {'Show': '#4e79a7', 'No Show': '#e15759'}
gender_colors = ['#ff9da7', '#86b4d7']  # Soft pink/blue for genders
# Create figure
plt.figure(figsize=(10, 6), facecolor='white')
# Set style
sns.set_style("whitegrid")
plt.rcParams['font.weight'] = 'bold'
# Bar positions
bar_width = 0.35
x = np.arange(len(gender))
# Plot bars
show_bars = plt.bar(x - bar_width/2, show, bar_width, 
                   label='Show', color=colors['Show'],
                   edgecolor='white', linewidth=2)
no_show_bars = plt.bar(x + bar_width/2, no_show, bar_width, 
                      label='No Show', color=colors['No Show'],
                      edgecolor='white', linewidth=2)
# Add counts on top of bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:,}',
                ha='center', va='bottom',
                fontsize=10, fontweight='bold')
add_labels(show_bars)
add_labels(no_show_bars)
# Customize appearance
plt.title('Appointment Attendance by Gender', 
          fontsize=14, pad=20, fontweight='bold')
plt.xlabel('Gender', fontsize=12, labelpad=10, fontweight='bold')
plt.ylabel('Number of Patients', fontsize=12, labelpad=10, fontweight='bold')
plt.xticks(x, gender, fontsize=11)
plt.yticks(fontsize=10)
# Add legend
legend = plt.legend(frameon=True, fontsize=11)
legend.get_frame().set_edgecolor('#dddddd')
# Add percentage annotations
total = np.array(show) + np.array(no_show)
show_percent = np.array(show)/total * 100
no_show_percent = np.array(no_show)/total * 100
for i in range(len(gender)):
    plt.text(x[i] - bar_width/2, show[i]/2, 
            f'{show_percent[i]:.1f}%', 
            ha='center', va='center',
            color='white', fontsize=10, fontweight='bold')
    plt.text(x[i] + bar_width/2, no_show[i]/2, 
            f'{no_show_percent[i]:.1f}%', 
            ha='center', va='center',
            color='white', fontsize=10, fontweight='bold')
# Remove spines
sns.despine(left=True)
plt.tight_layout()
plt.show()
```

```python
# Set style
sns.set_style("whitegrid")
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['axes.labelweight'] = 'bold'
# Sample data (replace with your actual data)
np.random.seed(42)
ages = np.concatenate([
    np.random.normal(40, 15, 2000),  # Showed up
    np.random.normal(50, 20, 500)    # No-show
])
status = ['Show']*2000 + ['No Show']*500
# Create dataframe
data = pd.DataFrame({'Age': ages, 'Status': status})
# Create figure
plt.figure(figsize=(10, 6), facecolor='white')
# Create histogram with hue
hist = sns.histplot(
    data=data,
    x='Age',
    hue='Status',
    bins=20,
    kde=True,
    palette=['#4e79a7', '#e15759'],  # Same colors as previous
    alpha=0.7,
    edgecolor='white',
    linewidth=1.5,
    hue_order=['Show', 'No Show']
)
# Customize appearance
plt.title('Age Distribution by No-Show Status', 
          fontsize=14, pad=20, fontweight='bold')
plt.xlabel('Age (Years)', fontsize=12, labelpad=10)
plt.ylabel('Number of Patients', fontsize=12, labelpad=10)
# Add mean age lines
show_mean = data[data['Status']=='Show']['Age'].mean()
no_show_mean = data[data['Status']=='No Show']['Age'].mean()
plt.axvline(show_mean, color='#4e79a7', linestyle='--', linewidth=2)
plt.axvline(no_show_mean, color='#e15759', linestyle='--', linewidth=2)
# Add mean annotations
plt.text(show_mean+2, plt.ylim()[1]*0.9, 
         f'Mean (Show): {show_mean:.1f} yrs',
         color='#4e79a7', fontweight='bold')
plt.text(no_show_mean+2, plt.ylim()[1]*0.8, 
         f'Mean (No Show): {no_show_mean:.1f} yrs',
         color='#e15759', fontweight='bold')
# Customize legend
legend = plt.legend(title='Status', frameon=True, fontsize=11)
legend.get_frame().set_edgecolor('#dddddd')
legend.get_title().set_fontweight('bold')
# Remove top/right spines
sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Set professional style
sns.set_style("whitegrid", {'grid.linestyle': '--', 'grid.alpha': 0.4})
plt.rcParams.update({
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold'
})
# Sample data (replace with your actual values)
ages = np.random.normal(loc=50, scale=20, size=2000)
ages = ages[(ages >= 0) & (ages <= 120)]  # Constrain to realistic ages
# Create figure
plt.figure(figsize=(10, 6), facecolor='white', dpi=100)
# Enhanced KDE plot
kde = sns.kdeplot(
    x=ages,
    color='#2b7bba',  # Professional blue
    fill=True,
    alpha=0.6,
    linewidth=2.5,
    edgecolor='white',
    label='Age Distribution'
)
# Customize axes
plt.title('Patient Age Distribution', fontsize=14, pad=20)
plt.xlabel('Age (Years)', fontsize=12, labelpad=10)
plt.ylabel('Density', fontsize=12, labelpad=10)
plt.xlim(0, 120)
plt.xticks(np.arange(0, 121, 20))
plt.ylim(0, 0.025)  # Adjusted to match your scale
# Add statistical markers
mean_age = np.mean(ages)
median_age = np.median(ages)
plt.axvline(mean_age, color='#d62728', linestyle='--', linewidth=2, label=f'Mean: {mean_age:.1f} yrs')
plt.axvline(median_age, color='#2ca02c', linestyle=':', linewidth=2, label=f'Median: {median_age:.1f} yrs')
# Add distribution info
plt.text(0.95, 0.95,
         f'Total Patients: {len(ages):,}\n'
         f'Std Dev: {np.std(ages):.1f} years',
         transform=plt.gca().transAxes,
         ha='right', va='top',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='#dddddd'))
# Custom legend
legend = plt.legend(frameon=True, loc='upper left')
legend.get_frame().set_edgecolor('#dddddd')
# Remove top/right spines
sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Set professional style
sns.set_style("whitegrid", {'grid.linestyle': '--', 'grid.alpha': 0.3})
plt.rcParams.update({
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold'
})
# Sample data (replace with your actual data)
data = pd.DataFrame({
    'gender': np.random.choice(['F', 'M'], 1000),
    'noshow': np.random.choice(['No', 'Yes'], 1000, p=[0.8, 0.2])
})
# Create figure
plt.figure(figsize=(10, 6), facecolor='white', dpi=100)
# Create countplot with custom styling
ax = sns.countplot(
    data=data,
    x='gender',
    hue='noshow',
    order=['F', 'M'],
    hue_order=['No', 'Yes'],
    palette=['#4e79a7', '#e15759'],  # Consistent colors
    saturation=0.9,
    edgecolor='white',
    linewidth=2
)
# Add percentage annotations
total_by_gender = data['gender'].value_counts()
for i, gender in enumerate(['F', 'M']):
    gender_data = data[data['gender'] == gender]
    no_show_percent = len(gender_data[gender_data['noshow'] == 'Yes']) / len(gender_data) * 100
    plt.text(i, len(gender_data)*0.9, 
             f'{no_show_percent:.1f}% No-Show', 
             ha='center', va='center',
             color='white', fontweight='bold',
             bbox=dict(facecolor='#333333', alpha=0.8, pad=3))
# Customize appearance
plt.title('No-Show Distribution by Gender', fontsize=14, pad=20)
plt.xlabel('Gender', fontsize=12, labelpad=10)
plt.ylabel('Number of Patients', fontsize=12, labelpad=10)
# Customize legend
legend = plt.legend(title='No-Show Status', frameon=True, 
                   labels=['Attended', 'No-Show'])
legend.get_frame().set_edgecolor('#dddddd')
legend.get_title().set_fontweight('bold')
# Add value labels on bars
for container in ax.containers:
    ax.bar_label(container, 
                label_type='edge', 
                padding=3,
                fontweight='bold',
                labels=[f'{x:,.0f}' for x in container.datavalues])
# Remove top/right spines
sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Set professional style
sns.set_style("whitegrid", {'grid.linestyle': ':', 'grid.alpha': 0.3})
plt.rcParams.update({
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold',
    'axes.titlepad': 20
})
# Sample data (replace with your actual data)
np.random.seed(42)
data = pd.DataFrame({
    'Age': np.concatenate([
        np.random.normal(45, 15, 600),  # Females
        np.random.normal(50, 20, 400)   # Males
    ]),
    'Gender': ['F']*600 + ['M']*400
})
data = data[(data['Age'] >= 0) & (data['Age'] <= 100)]  # Filter realistic ages
# Create figure
plt.figure(figsize=(12, 6), facecolor='white', dpi=100)
# Create enhanced histogram
hist = sns.histplot(
    data=data,
    x='Age',
    hue='Gender',
    bins=20,
    kde=True,
    palette=['#ff9da7', '#86b4d7'],  # Soft pink/blue
    alpha=0.7,
    edgecolor='white',
    linewidth=1.5,
    hue_order=['F', 'M'],
    stat='density'
)
# Customize appearance
plt.title('Age Distribution by Gender (Attended Patients)', fontsize=14)
plt.xlabel('Age (Years)', fontsize=12, labelpad=10)
plt.ylabel('Density', fontsize=12, labelpad=10)
plt.xlim(0, 100)
plt.xticks(np.arange(0, 101, 10))
# Add mean age lines
for gender, color in zip(['F', 'M'], ['#ff6b81', '#5a9bd5']):
    mean_age = data[data['Gender'] == gender]['Age'].mean()
    plt.axvline(mean_age, color=color, linestyle='--', linewidth=2.5)
    plt.text(mean_age+1, plt.ylim()[1]*0.9-(0.05 if gender=='M' else 0), 
             f'{gender} Mean: {mean_age:.1f} yrs',
             color=color, fontweight='bold')
# Customize legend
legend = plt.legend(title='Gender', labels=['Female', 'Male'], frameon=True)
legend.get_frame().set_edgecolor('#dddddd')
legend.get_title().set_fontweight('bold')
# Add distribution info
plt.text(0.95, 0.95,
         f"Total Patients: {len(data):,}\n"
         f"Female: {len(data[data['Gender']=='F']):,} ({len(data[data['Gender']=='F'])/len(data)*100:.1f}%)\n"
         f"Male: {len(data[data['Gender']=='M']):,} ({len(data[data['Gender']=='M'])/len(data)*100:.1f}%)",
         transform=plt.gca().transAxes,
         ha='right', va='top',
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='#dddddd', pad=5),
         fontsize=10)
# Remove top/right spines
sns.despine()
plt.tight_layout()
plt.show()
```
